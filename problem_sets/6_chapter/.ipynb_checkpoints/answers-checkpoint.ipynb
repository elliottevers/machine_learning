## Problem 7

![alttext](img/p7a.png)
![alttext](img/p7b.png)

#### (a)

**a single residual**

The $i$-th residual is
$$
\big[ Y _ { i } - \left( \beta _ { 0 } + \sum _ { j = 1 } ^ { p } \beta _ { j } X _ { i j } \right) \big]^2
$$.

**probability of a datum**

Since we've assumed the residuals are normally distributed, the probability of seeing the $i$-th datum is
$$
\frac { 1 } { \sqrt { 2 \pi \sigma ^ { 2 } } } e ^ { - \frac { \big[ Y _ { i } - \left( \beta _ { 0 } + \sum _ { j = 1 } ^ { p } \beta _ { j } X _ { i j } \right) \big]^2 } { 2 \sigma ^ { 2 } } }
$$.

**probability of data (product of those probabilities)**

This means the probability of seeing *all* the data is
$$
\prod _ { i = 1 } ^ { n } \frac { 1 } { \sqrt { 2 \pi \sigma ^ { 2 } } } e ^ { - \frac { \left[ Y _ { i } - \left( \beta _ { 0 } + \sum _ { j = 1 } ^ { p } \beta _ { j } X _ { i j } \right) \right] ^ { 2 } } { 2 \sigma ^ { 2 } } }
$$.

**likelihood, which is a function of distribution parameters ($\beta$ in this case) that is *also* a probability distribution (simplify)**

Simplifying, we get the likehood of
$$
\mathcal{L} ( \theta | \beta ) = \\ p(\beta | \theta) = \\ \left( \frac { 1 } { \sigma \sqrt { 2 \pi } } \right) ^ { n } e^{ \left( - \frac { 1 } { 2 \sigma ^ { 2 } } \sum _ { i = 1 } ^ { n } \left[ Y _ { i } - \left( \beta _ { 0 } + \sum _ { j = 1 } ^ { p } \beta _ { j } X _ { i j } \right) \right] ^ { 2 } \right)}
$$.

#### (b)

**probability of single beta (laplacian)**

The density of the $i$-th $\beta$ is
$$
p \left( \beta _ { i } \right) = \frac { 1 } { 2 b } e^{ \left( - \frac { \left| \beta _ { i } \right| } { b } \right) }
$$

**full prior (product of laplacians)**

Since we have $p$ predictors, the probability for *all* the betas for a given fit is 
$$
p(\beta) = \frac { 1 } { ( 2 b ) ^ { p } } e^{ \left( - \frac { 1 } { b } \sum _ { j = 1 } ^ { p } \left| \beta _ { i } \right| \right)}
$$

**posterior is a product of likelihood and prior**

Finally, we have our posterior
$$
p(\theta | \beta) = \mathcal { L } ( \theta | \beta ) p ( \beta ) =
\frac { 1 } { ( 2 b ) ^ { p } ( 2 \pi ) ^ { n / 2 } \sigma ^ { n } } e ^{  - \frac { 1 } { 2 \sigma ^ { 2 } } \sum _ { i = 1 } ^ { n } \left( y _ { i } - \beta _ { 0 } - \sum _ { j = 1 } ^ { p } x _ { i j } \beta _ { j } \right) ^ { 2 } - \frac { 1 } { b } \sum _ { j = 1 } ^ { p } \left| \beta _ { j } \right|}
$$

#### (c)

The mode of a (unimodal) distribution is where it achieves its maximum, so we can phrase our optimization problem as 
$$
\operatorname*{argmax}_\beta p ( \theta | \beta )
$$.

We can make our optimization problem a little easier to handle by getting rid of the exponential
$$
\underset { \beta } { \operatorname { argmax } } \log \left[ \left( \frac { 1 } { \sigma \sqrt { 2 \pi } } \right) ^ { n } \left( \frac { 1 } { 2 b } \right) \right] - \left( \frac { 1 } { 2 \sigma ^ { 2 } } \sum _ { i = 1 } ^ { n } \left[ Y _ { i } - \left( \beta _ { 0 } + \sum _ { j = 1 } ^ { p } \beta _ { j } X _ { i j } \right) \right] ^ { 2 } + \frac { | \beta | } { b } \right)
$$.

Since the logarithmic function is monotonically increasing, minimizing this new transformation is equivalent to minimizing the first.  Furthermore, since $\left[ \left( \frac { 1 } { \sigma \sqrt { 2 \pi } } \right) ^ { n } \left( \frac { 1 } { 2 b } \right) \right]$ is a constant, this simpler maximization problem amounts to the even simpler minimization problem
$$
\underset { \beta } { \operatorname { argmin } } \frac { 1 } { 2 \sigma ^ { 2 } } \sum _ { i = 1 } ^ { n } \left[ Y _ { i } - \left( \beta _ { 0 } + \sum _ { j = 1 } ^ { p } \beta _ { j } X _ { i j } \right) \right] ^ { 2 } + \frac { | \beta | } { b }
$$.

This easily simplifies to exactly the minimization for the LASSO
$$
\underset { \beta } { \operatorname { argmin } } \sum _ { i = 1 } ^ { n } \left( y _ { i } - \beta _ { 0 } - \sum _ { j = 1 } ^ { p } \beta _ { j } x _ { i j } \right) ^ { 2 } + \lambda \sum _ { j = 1 } ^ { p } | \beta _ { j } |
$$.

## Problem 8

![alttext](img/p8a.png)
![alttext](img/p8b.png)

#### (a)

```r
set.seed(9002)
X = rnorm(100)
eps = rnorm(100)
```

#### (b)

**simulate data**

```r
beta_0 = 1
beta_1 = 2
beta_2 = 4
beta_3 = 8
Y = beta_0 + beta_1 * X + beta_2 * X^2 + beta_3 * X^3 + eps
```

#### (c)

```r
install.packages('leaps')
library(leaps)
```

**For each subset size of predictor and criterion, produce a score**

```r
data = data.frame(
    y = Y,
    x = X
)

models = regsubsets(
    y ~ poly(
        x,
        10,
        raw = T
    ),
    data = data,
    nvmax = 10
)

summary = summary(
    models
)
```

**for each model, find minimum score**

```r
plot(
    summary$cp,
    xlab = "Number of Predictors",
    ylab = "Cp",
    type = "l"
)

cp_min = which.min(
    summary$cp
)

points(
    cp_min,
    summary$cp[cp_min],
    col = "red",
    lwd = 10
)
```

```r
plot(
    mod.summary$bic,
    xlab = "Number of Predictors",
    ylab = "BIC",
    type = "l"
)

bic_min = which.min(
    summary$bic
)

points(
    bic_min,
    mod.summary$bic[bic_min],
    col = "red",
    lwd = 10
)
```

```r
plot(
    summary$adjr2,
    xlab = "Number of Predictors",
    ylab = "Adjusted R2",
    type = "l"
)

adjr2_max = which.max(
    summary$adjr2
)

points(
    adjr2_max,
    summary$adjr2[adjr2_max],
    col = "red",
    lwd = 10
)
```

**Given a minimum score, let R choose best model**

```r
coefficients(
    models,
    id = 3
)
```
