Recall the form of a boosted model
$$\hat { f } ( x ) = \sum _ { b = 1 } ^ { B } \lambda \hat { f } ^ { b } ( x )$$.
If $B = p$ stumps chooses $B$ distinct predictors to split on, then the answer is obvious, i.e., let $f_j = f_b = \lambda \hat{f}^{b}$.

But we don't have to have a set of distinct predictors that our stumps are splitting on.  If we never choose one of the predictors, then we have the $0$ function which we are adding to the model.  Let $i \in P$ denote the $i$th stump splits on the predictor $p$.  If $|P| > 1$, i.e. we choose the predictor multiple times, then this *adds* a contribution of $f_j = \sum _ { b \in P } \lambda \hat { f } ^ { b } ( x )$ to our boosted model.

```r
# proportion of training observations in the m-th region from the k-th class
p_mk = seq(0, 1, 0.01)

gini_index = p_mk * (1 - p_mk) + (1 - p_mk) * p_mk

E = 1 - pmax(p_mk, 1 - p_mk)

cross_entropy = -1 * (
    (p_mk * log(p_mk)) + ((1 - p_mk) * log(1 - p_mk))
)

plot(
    p_mk,
    gini_index,
    col = "red"
)

par(new = TRUE)

plot(
    p_mk,
    E,
    col = "green"
)

par(new = TRUE)

plot(
    p_mk,
    cross_entropy,
    col = "blue"
)
```

## Problem 7

![alttext](img/p7.png)

```r
install.packages('randomForest')
```

```r
library(MASS)
library(randomForest)
```

```r
set.seed(9001)

params_trees = 1:500

data = Boston

# sample 
i_train = sample(
    dim(Boston)[1],
    dim(Boston)[1]/2
)

X_train = data[i_train, -14]
X_test = data[-i_train, -14]
Y_train = data[i_train, 14]
Y_test = data[-i_train, 14]

p = dim(data)[2] - 1

p_id = p
p_halve = p/2
p_sqrt = sqrt(p)

model_p_id = randomForest(
    X_train,
    Y_train,
    xtest = X_test,
    ytest = Y_test, 
    mtry = p_id,
    ntree = 500
)

model_p_halve = randomForest(
    X_train,
    Y_train,
    xtest = X_test,
    ytest = Y_test, 
    mtry = p_halve,
    ntree = 500
)

model_p_sqrt = randomForest(
    X_train,
    Y_train,
    xtest = X_test,
    ytest = Y_test, 
    mtry = p_sqrt,
    ntree = 500
)

#plots

plot(
    params_trees,
    model_p_id$test$mse,
    col = "red",
    type = "l",
    xlab = "Number of Trees", 
    ylab = "Test Classification Error"
)

lines(
    params_trees,
    model_p_halve$test$mse,
    col = "blue",
    type = "l"
)

lines(
    params_trees,
    model_p_sqrt$test$mse,
    col = "green",
    type = "l"
)

legend(
    "topright",
    c("m=p", "m=p/2", "m=sqrt(p)"), 
    col = c("red", "blue", "green"), 
    cex = 1,
    lty = 1
)
```

Halving the number of predictors to randomly choose from produced the most benefit.

## Problem 8

![alttext](img/p8.png)

## (a)

```r
library(ISLR)
set.seed(9001)

data = Carseats

i_train = sample(
    dim(data)[1],
    dim(data)[1]/2
)

data_train = data[i_train, ]
data_test = data[-i_train, ]
```

## (b)

```r
install.packages('tree')
```

```r
library(tree)

model = tree(
    Sales ~ .,
    data = data_train
)

summary(model)

plot(model)

text(model)

y_hat = predict(
    model,
    data_test
)

mean(
    (data_test$Sales - y_hat)^2
)
```

**test error is 4.33**

## (c)

```r
cv_model = cv.tree(
    model,
    FUN = prune.tree
)

plot(
    cv_model$size, # size of subtreeb
    cv_model$dev, # cost
    type = "b"
)

plot(
    cv_model$k, # hyperparameter for penalty for subtree complexity
    cv_model$dev, # cost
    type = "b"
)
```

```r
model_pruned = prune.tree(
    model,
    best = 14
)

plot(model_pruned)

text(model_pruned)
```

```r
y_hat = predict(
    model_pruned,
    data_test
)

mean(
    (data_test$Sales - y_hat)^2
)
```

**pruning the model actually decreased the test error**

## (d - e)

```r
library(randomForest)

model_rf = randomForest(
    Sales ~ .,
    data = data_train,
    mtry = 10,
    ntree = 500, 
    importance = T
)

y_hat_rf = predict(
    model_rf,
    data_test
)

mean(
    (data_test$Sales - y_hat_rf)^2
)

importance(model_rf)
```

**using random forests improves the test error**
