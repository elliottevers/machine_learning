Recall the form of a boosted model
$$\hat { f } ( x ) = \sum _ { b = 1 } ^ { B } \lambda \hat { f } ^ { b } ( x )$$.
If $B = p$ stumps chooses $B$ distinct predictors to split on, then the answer is obvious, i.e., let $f_j = f_b = \lambda \hat{f}^{b}$.

But we don't have to have a set of distinct predictors that our stumps are splitting on.  If we never choose one of the predictors, then we have the $0$ function which we are adding to the model.  Let $i \in P$ denote the $i$th stump splits on the predictor $p$.  If $|P| > 1$, i.e. we choose the predictor multiple times, then this *adds* a contribution of $f_j = \sum _ { b \in P } \lambda \hat { f } ^ { b } ( x )$ to our boosted model.

```r
# proportion of training observations in the m-th region from the k-th class
p_mk = seq(0, 1, 0.01)

gini_index = p_mk * (1 - p_mk) + (1 - p_mk) * p_mk

E = 1 - pmax(p_mk, 1 - p_mk)

cross_entropy = -1 * (
    (p_mk * log(p_mk)) + ((1 - p_mk) * log(1 - p_mk))
)

plot(
    p_mk,
    gini_index,
    col = "red"
)

par(new = TRUE)

plot(
    p_mk,
    E,
    col = "green"
)

par(new = TRUE)

plot(
    p_mk,
    cross_entropy,
    col = "blue"
)
```
